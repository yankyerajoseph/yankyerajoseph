```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
import io
import math
from google.colab import files

# Your custom equation
def Model_validation_fit(t, A=556.41, mu=90.87, lag=0.48):
    """
    Custom biogas production model.

    Parameters:
    t (array-like): Time values.
    A (float): Maximum biogas potential.
    mu (float): Rate constant.
    lag (float): Lag phase duration.

    Returns:
    array-like: Biogas yield at time t.
    """
    return A * np.exp(-np.exp((mu * np.exp(1) / A) * (lag - t) + 1))

# Assuming 'data' is your DataFrame containing 'HRT (d)' and 'NHMM (ml/g oTS)'
t_data = data[' HRT (d) '].values
B_data = data[' NHMM (ml/g oTS) '].values
# Define the parameters explicitly
A = 556.41  # Maximum biogas potential
mu = 90.87  # Rate constant
lag = 0.48  # Lag phase duratio

# Generate predictions using your custom model
B_pred = Model_validation_fit (t_data)

# Evaluate the model
r2 = r2_score(B_data, B_pred.round(2))
rmse = np.sqrt(mean_squared_error(B_data, B_pred.round(2)))

# Print the evaluation metrics
print(f"R-squared: {r2}")
print(f"RMSE: {rmse}")

# Plot the results
plt.plot(t_data, B_data, 'o', label='Observed Data')
plt.plot(t_data, B_pred, '-', label='Model Prediction')
plt.xlabel('HRT (d)')
plt.ylabel('NHMM (ml/g oTS)')
plt.legend()
plt.title('Model_validation_fit')
plt.show()

# DataFrame
predicted_data = pd.DataFrame({'HRT (d)': t_data, 'NHMM (ml/g oTS)': B_pred.round(2)})

# Predicted data in pairs
display(predicted_data)

print ('Model_validation_fit')

# Memory buffer
buffer = io.BytesIO()

#Save DataFrame
predicted_data.to_csv('predicted_data.csv', index=False)

# Rewind buffer to the start
buffer.seek(0)

# Downloading buffer
files.download('predicted_data.csv')

print(f"Equation: B(t) = {A:.2f} * exp(-exp(({mu:.2f} * {math.e:.2f} / {A:.2f}) * ({lag:.2f} - t) + 1))")
```

```python
# Cross Validation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold

# Function to calculate and print Adjusted R-squared
def calculate_adj_r2(y_true, x):
    X = sm.add_constant(x)
    model = sm.OLS(y_true, X).fit()
    adj_r2 = model.rsquared_adj
    return adj_r2

# Function to calculate MAE
def calculate_mae(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    return mae

# Number of folds for cross-validation
n_splits = 5  # You can adjust this value

# Define the custom_model function (Modified Gompertz JY Model)
def custom_model(t):
    return 556.41 * np.exp(-np.exp((90.87 * np.e / 556.41) * (0.48 - t) + 1))

# Perform cross-validation for each sample
def cross_validate(data, target_col):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)  # Create KFold object
    mae_scores = []
    adj_r2_scores = []

    for train_index, test_index in kf.split(data):
        X_train, X_test = data[' HRT (d) '].values[train_index], data[' HRT (d) '].values[test_index]
        y_train, y_test = data[target_col].values[train_index], data[target_col].values[test_index]

        # Fit the model on the training data
        B_pred_train = custom_model(X_train)

        # Predict on the test data
        B_pred_test = custom_model(X_test)

        # Calculate metrics for this fold
        mae_scores.append(calculate_mae(y_test, B_pred_test))
        adj_r2_scores.append(calculate_adj_r2(y_test, X_test))

    return np.mean(mae_scores), np.mean(adj_r2_scores)

# Cross-validation for each sample
mae_NHMM, adj_r2_NHMM = cross_validate(data, ' NHMM (ml/g oTS) ')

# Print the cross-validation results
print(f"NHMM - Cross-validated MAE: {mae_NHMM}, Adjusted R-squared: {adj_r2_NHMM}")
```

```python
# Sensitivity analysis: Cross-validation:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold

# Test the Model_validation function with some example input
example_input = np.array([10, 20, 30])  # Example HRT values
output = Model_validation_fit(example_input)  # Call the custom_model function
print(f"Output from custom_model: {output}")  # Print the output

# Test the cross_validate function with a single fold and default model
mae, adj_r2 = cross_validate(data, ' NHMM (ml/g oTS) ')  # Use default model_func
print(f"MAE: {mae}, Adjusted R-squared: {adj_r2}")

# Define parameter ranges for sensitivity analysis
param_ranges = {
    'A': np.linspace(500, 600, 10),
    'mu': np.linspace(80, 100, 10),
    'lag': np.linspace(0.4, 0.6, 10)
}

# Perform sensitivity analysis with cross-validation
results = []
for A in param_ranges['A']:
    for mu in param_ranges['mu']:
        for lag in param_ranges['lag']:
            # Modify Model_validation with current parameter values
            def Model_validation_with_params(t, A=A, mu=mu, lag=lag): #Added default parameters here
                return Model_validation_fit(t, A=A, mu=mu, lag=lag) #Calling Model_validation_fit

            # Perform cross-validation, passing custom_model_with_params
            mae_NHMM, adj_r2_NHMM = cross_validate(data, ' NHMM (ml/g oTS) ', model_func=Model_validation_with_params) #Referencing the correct name
            results.append([A, mu, lag, mae_NHMM, adj_r2_NHMM])
```

```python
# Plotting the bootstrap distributions and confidence intervals

def plot_bootstrap_results(data, metric_name, sample_name):
    plt.figure(figsize=(8, 6))
    sns.histplot(data, kde=True)  # Use seaborn for histogram with KDE
    plt.title(f"Bootstrap Distribution of {metric_name} for {sample_name}")
    plt.xlabel(metric_name)
    plt.ylabel("Frequency")

    # Add confidence interval lines
    ci_lower, ci_upper = confidence_interval(data)
    plt.axvline(ci_lower, color='r', linestyle='--', label=f"Lower CI ({ci_lower:.2f})")
    plt.axvline(ci_upper, color='r', linestyle='--', label=f"Upper CI ({ci_upper:.2f})")

    plt.legend()
    plt.show()

# Plot for MAE
plot_bootstrap_results(mae_NHMM_bootstrap, "MAE", "NHMM")


# Plot for Adjusted R-squared
plot_bootstrap_results(adj_r2_NHMM_bootstrap, "Adjusted R-squared", "NHMM")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import io
from google.colab import files
from scipy.integrate import odeint
from scipy.optimize import curve_fit
import statsmodels.api as sm # Import statsmodels for Adjusted R-squared

# First order differential equation
def biogas_rate_first_order(B, t, k, B_max):
    dBdt = k * (B_max - B)
    return dBdt

# Second-order differential equation
def biogas_rate_second_order(y, t, k1, k2, B_max):
    B, dBdt = y
    d2Bdt2 = k1 * (B_max - B) - k2 * dBdt
    return [dBdt, d2Bdt2]

# Modified Gompertz JY Model
def B_gompertz(t, A, mu, lag):
    return A * np.exp(-np.exp((mu * np.e / A) * (lag - t) + 1))


# Initial conditions for first-order model
B0_first_order = 0

# Initial conditions for second-order model
B0_second_order = 0
dBdt0_second_order = 0
y0_second_order = [B0_second_order, dBdt0_second_order]

# Assuming 'data' is your DataFrame containing 'HRT (d)', 'HM (ml/g oTS)', and ' NHMM (ml/g oTS)'
t_data = data[' HRT (d) '].values
hm_data = data[' HM (ml/g oTS) '].values
nhmm_data = data[' NHMM (ml/g oTS) '].values

# --- Fitting and Predicting for HM ---

# Fit the first-order model to HM data
initial_guess_first_order_hm = [0.1, 500]
popt_first_order_hm, pcov_first_order_hm = curve_fit(
    lambda t, k, B_max: odeint(biogas_rate_first_order, B0_first_order, t, args=(k, B_max))[:, 0],
    t_data, hm_data,
    p0=initial_guess_first_order_hm
)
k_fit_first_order_hm, B_max_fit_first_order_hm = popt_first_order_hm
B_pred_first_order_hm = odeint(biogas_rate_first_order, B0_first_order, t_data, args=(k_fit_first_order_hm, B_max_fit_first_order_hm))[:, 0]

# Fit the second-order model to HM data
initial_guess_second_order_hm = [0.1, 0.1, 500]
popt_second_order_hm, pcov_second_order_hm = curve_fit(
    lambda t, k1, k2, B_max: odeint(biogas_rate_second_order, y0_second_order, t, args=(k1, k2, B_max))[:, 0],
    t_data, hm_data,
    p0=initial_guess_second_order_hm
)
k1_fit_second_order_hm, k2_fit_second_order_hm, B_max_fit_second_order_hm = popt_second_order_hm
B_pred_second_order_hm = odeint(biogas_rate_second_order, y0_second_order, t_data, args=(k1_fit_second_order_hm, k2_fit_second_order_hm, B_max_fit_second_order_hm))[:, 0]

# Fit the Modified Gompertz model to HM data
# You need to provide initial guesses for A, mu, lag for the Gompertz model on HM
initial_guess_gompertz_hm = [500, 80, 0.5]
popt_gompertz_hm, pcov_gompertz_hm = curve_fit(
    B_gompertz,
    t_data, hm_data,
    p0=initial_guess_gompertz_hm
)
A_fit_gompertz_hm, mu_fit_gompertz_hm, lag_fit_gompertz_hm = popt_gompertz_hm
B_pred_gompertz_hm = B_gompertz(t_data, A_fit_gompertz_hm, mu_fit_gompertz_hm, lag_fit_gompertz_hm)


# --- Fitting and Predicting for NHMM ---

# Fit the Modified Gompertz model to NHMM data
# You need to provide initial guesses for A, mu, lag for the Gompertz model on NHMM
# Based on your previous code, you used A=556.41, mu=90.87, lag=0.48. You could use these as initial guesses
initial_guess_gompertz_nhmm = [550, 90, 0.5] # Using slightly different initial guesses for demonstration
popt_gompertz_nhmm, pcov_gompertz_nhmm = curve_fit(
    B_gompertz,
    t_data, nhmm_data,
    p0=initial_guess_gompertz_nhmm
)
A_fit_gompertz_nhmm, mu_fit_gompertz_nhmm, lag_fit_gompertz_nhmm = popt_gompertz_nhmm
B_pred_gompertz_nhmm = B_gompertz(t_data, A_fit_gompertz_nhmm, mu_fit_gompertz_nhmm, lag_fit_gompertz_nhmm)


# --- Function to calculate Adjusted R-squared ---
def calculate_adj_r2(y_true, y_pred, num_params):
    # For OLS model, we need the predictor (which is implicit in time series models)
    # A simplified way for non-linear models is to use the definition:
    # Adj R^2 = 1 - [(1 - R^2) * (n - 1) / (n - k - 1)]
    # where n is the number of data points, k is the number of parameters
    r2 = r2_score(y_true, y_pred)
    n = len(y_true)
    k = num_params
    if n - k - 1 <= 0:
        return np.nan # Avoid division by zero or negative
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
    return adj_r2

# --- Calculate and Print Evaluation Metrics for Each Model ---

print("--- Model Evaluation Results ---")

# First-Order Differential Model (HM)
r2_first_order_hm = r2_score(hm_data, B_pred_first_order_hm)
rmse_first_order_hm = np.sqrt(mean_squared_error(hm_data, B_pred_first_order_hm))
mae_first_order_hm = mean_absolute_error(hm_data, B_pred_first_order_hm)
adj_r2_first_order_hm = calculate_adj_r2(hm_data, B_pred_first_order_hm, num_params=2) # 2 parameters: k, B_max
print("\nFirst-Order Differential Model (HM):")
print(f"  Fitted Parameters: k={k_fit_first_order_hm:.4f}, B_max={B_max_fit_first_order_hm:.4f}")
print(f"  R-squared: {r2_first_order_hm:.4f}")
print(f"  Adjusted R-squared: {adj_r2_first_order_hm:.4f}")
print(f"  RMSE: {rmse_first_order_hm:.4f}")
print(f"  MAE: {mae_first_order_hm:.4f}")

# Second-Order Differential Model (HM)
r2_second_order_hm = r2_score(hm_data, B_pred_second_order_hm)
rmse_second_order_hm = np.sqrt(mean_squared_error(hm_data, B_pred_second_order_hm))
mae_second_order_hm = mean_absolute_error(hm_data, B_pred_second_order_hm)
adj_r2_second_order_hm = calculate_adj_r2(hm_data, B_pred_second_order_hm, num_params=3) # 3 parameters: k1, k2, B_max
print("\nSecond-Order Differential Model (HM):")
print(f"  Fitted Parameters: k1={k1_fit_second_order_hm:.4f}, k2={k2_fit_second_order_hm:.4f}, B_max={B_max_fit_second_order_hm:.4f}")
print(f"  R-squared: {r2_second_order_hm:.4f}")
print(f"  Adjusted R-squared: {adj_r2_second_order_hm:.4f}")
print(f"  RMSE: {rmse_second_order_hm:.4f}")
print(f"  MAE: {mae_second_order_hm:.4f}")

# Modified Gompertz JY Model (HM)
r2_gompertz_hm = r2_score(hm_data, B_pred_gompertz_hm)
rmse_gompertz_hm = np.sqrt(mean_squared_error(hm_data, B_pred_gompertz_hm))
mae_gompertz_hm = mean_absolute_error(hm_data, B_pred_gompertz_hm)
adj_r2_gompertz_hm = calculate_adj_r2(hm_data, B_pred_gompertz_hm, num_params=3) # 3 parameters: A, mu, lag
print("\nModified Gompertz JY Model (HM - Fitted):")
print(f"  Fitted Parameters: A={A_fit_gompertz_hm:.4f}, mu={mu_fit_gompertz_hm:.4f}, lag={lag_fit_gompertz_hm:.4f}")
print(f"  R-squared: {r2_gompertz_hm:.4f}")
print(f"  Adjusted R-squared: {adj_r2_gompertz_hm:.4f}")
print(f"  RMSE: {rmse_gompertz_hm:.4f}")
print(f"  MAE: {mae_gompertz_hm:.4f}")

# Modified Gompertz JY Model (NHMM - Fitted)
r2_gompertz_nhmm = r2_score(nhmm_data, B_pred_gompertz_nhmm)
rmse_gompertz_nhmm = np.sqrt(mean_squared_error(nhmm_data, B_pred_gompertz_nhmm))
mae_gompertz_nhmm = mean_absolute_error(nhmm_data, B_pred_gompertz_nhmm)
adj_r2_gompertz_nhmm = calculate_adj_r2(nhmm_data, B_pred_gompertz_nhmm, num_params=3) # 3 parameters: A, mu, lag
print("\nModified Gompertz JY Model (NHMM - Fitted):")
print(f"  Fitted Parameters: A={A_fit_gompertz_nhmm:.4f}, mu={mu_fit_gompertz_nhmm:.4f}, lag={lag_fit_gompertz_nhmm:.4f}")
print(f"  R-squared: {r2_gompertz_nhmm:.4f}")
print(f"  Adjusted R-squared: {adj_r2_gompertz_nhmm:.4f}")
print(f"  RMSE: {rmse_gompertz_nhmm:.4f}")
print(f"  MAE: {mae_gompertz_nhmm:.4f}")

# --- Plotting the results (as in your original code) ---
plt.figure(figsize=(12, 8))

plt.plot(t_data, hm_data, 'o', label='Experimental Data of HM')
plt.plot(t_data, B_pred_first_order_hm, '-', label='First-Order Differential Model Prediction of HM')
plt.plot(t_data, B_pred_second_order_hm, '-', label='Second-Order Differential Model Prediction of HM')
plt.plot(t_data, B_pred_gompertz_hm, '-', label='Modified Gompertz JY Model (HM)')

plt.plot(t_data, nhmm_data, 'x', label='Experimental Data of NHMM')
plt.plot(t_data, B_pred_gompertz_nhmm, '--', label='Modified Gompertz JY Model (NHMM)')

plt.xlabel('HRT (d)')
plt.ylabel('Biogas Yield (ml/g oTS)')
plt.title('Model Comparison as means of Validating the Modifed Gompertz JY model')
plt.legend()
plt.grid(False)
plt.show()
```
