```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import r2_score, mean_squared_error
import io
import math
from google.colab import files

# Your custom equation
def Model_validation_fit(t, A=556.41, mu=90.87, lag=0.48):
    """
    Custom biogas production model.

    Parameters:
    t (array-like): Time values.
    A (float): Maximum biogas potential.
    mu (float): Rate constant.
    lag (float): Lag phase duration.

    Returns:
    array-like: Biogas yield at time t.
    """
    return A * np.exp(-np.exp((mu * np.exp(1) / A) * (lag - t) + 1))

# Assuming 'data' is your DataFrame containing 'HRT (d)' and 'NHMM (ml/g oTS)'
t_data = data[' HRT (d) '].values
B_data = data[' NHMM (ml/g oTS) '].values
# Define the parameters explicitly
A = 556.41  # Maximum biogas potential
mu = 90.87  # Rate constant
lag = 0.48  # Lag phase duratio

# Generate predictions using your custom model
B_pred = Model_validation_fit (t_data)

# Evaluate the model
r2 = r2_score(B_data, B_pred.round(2))
rmse = np.sqrt(mean_squared_error(B_data, B_pred.round(2)))

# Print the evaluation metrics
print(f"R-squared: {r2}")
print(f"RMSE: {rmse}")

# Plot the results
plt.plot(t_data, B_data, 'o', label='Observed Data')
plt.plot(t_data, B_pred, '-', label='Model Prediction')
plt.xlabel('HRT (d)')
plt.ylabel('NHMM (ml/g oTS)')
plt.legend()
plt.title('Model_validation_fit')
plt.show()

# DataFrame
predicted_data = pd.DataFrame({'HRT (d)': t_data, 'NHMM (ml/g oTS)': B_pred.round(2)})

# Predicted data in pairs
display(predicted_data)

print ('Model_validation_fit')

# Memory buffer
buffer = io.BytesIO()

#Save DataFrame
predicted_data.to_csv('predicted_data.csv', index=False)

# Rewind buffer to the start
buffer.seek(0)

# Downloading buffer
files.download('predicted_data.csv')

print(f"Equation: B(t) = {A:.2f} * exp(-exp(({mu:.2f} * {math.e:.2f} / {A:.2f}) * ({lag:.2f} - t) + 1))")
```

```python
# Cross Validation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold

# Function to calculate and print Adjusted R-squared
def calculate_adj_r2(y_true, x):
    X = sm.add_constant(x)
    model = sm.OLS(y_true, X).fit()
    adj_r2 = model.rsquared_adj
    return adj_r2

# Function to calculate MAE
def calculate_mae(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    return mae

# Number of folds for cross-validation
n_splits = 5  # You can adjust this value

# Define the custom_model function (Modified Gompertz JY Model)
def custom_model(t):
    return 556.41 * np.exp(-np.exp((90.87 * np.e / 556.41) * (0.48 - t) + 1))

# Perform cross-validation for each sample
def cross_validate(data, target_col):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)  # Create KFold object
    mae_scores = []
    adj_r2_scores = []

    for train_index, test_index in kf.split(data):
        X_train, X_test = data[' HRT (d) '].values[train_index], data[' HRT (d) '].values[test_index]
        y_train, y_test = data[target_col].values[train_index], data[target_col].values[test_index]

        # Fit the model on the training data
        B_pred_train = custom_model(X_train)

        # Predict on the test data
        B_pred_test = custom_model(X_test)

        # Calculate metrics for this fold
        mae_scores.append(calculate_mae(y_test, B_pred_test))
        adj_r2_scores.append(calculate_adj_r2(y_test, X_test))

    return np.mean(mae_scores), np.mean(adj_r2_scores)

# Cross-validation for each sample
mae_NHMM, adj_r2_NHMM = cross_validate(data, ' NHMM (ml/g oTS) ')

# Print the cross-validation results
print(f"NHMM - Cross-validated MAE: {mae_NHMM}, Adjusted R-squared: {adj_r2_NHMM}")
```

```python
# Sensitivity analysis: Cross-validation:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold

# Test the Model_validation function with some example input
example_input = np.array([10, 20, 30])  # Example HRT values
output = Model_validation_fit(example_input)  # Call the custom_model function
print(f"Output from custom_model: {output}")  # Print the output

# Test the cross_validate function with a single fold and default model
mae, adj_r2 = cross_validate(data, ' NHMM (ml/g oTS) ')  # Use default model_func
print(f"MAE: {mae}, Adjusted R-squared: {adj_r2}")

# Define parameter ranges for sensitivity analysis
param_ranges = {
    'A': np.linspace(500, 600, 10),
    'mu': np.linspace(80, 100, 10),
    'lag': np.linspace(0.4, 0.6, 10)
}

# Perform sensitivity analysis with cross-validation
results = []
for A in param_ranges['A']:
    for mu in param_ranges['mu']:
        for lag in param_ranges['lag']:
            # Modify Model_validation with current parameter values
            def Model_validation_with_params(t, A=A, mu=mu, lag=lag): #Added default parameters here
                return Model_validation_fit(t, A=A, mu=mu, lag=lag) #Calling Model_validation_fit

            # Perform cross-validation, passing custom_model_with_params
            mae_NHMM, adj_r2_NHMM = cross_validate(data, ' NHMM (ml/g oTS) ', model_func=Model_validation_with_params) #Referencing the correct name
            results.append([A, mu, lag, mae_NHMM, adj_r2_NHMM])
```
